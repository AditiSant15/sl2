```python
#step function
def step_function(y_in,theta):
    if y_in >= theta:
        return 1
    else:
        return 0
#input all comnination
inputs =[(0,0),(0,1),(1,0),(1,1)]

#weights and threshould
w1=1
w2=-1
theta=1
print("X1  X2  Y ")
#comput output
for x1,x2 in inputs:
    y_in=w1*x1+w2*x2
    y=step_function(y_in,theta)
    print(x1," ",x2," ",y," ")
```

```python
# McCulloch-Pitts ANDNOT Function with User Input

def mcculloch_pitts_andnot(x1, x2):
    w1 = 1
    w2 = -1
    theta = 1

    net_input = (w1 * x1) + (w2 * x2)

    if net_input >= theta:
        return 1
    else:
        return 0

# Taking input from user
x1 = int(input("Enter value of x1 (0 or 1): "))
x2 = int(input("Enter value of x2 (0 or 1): "))

# Validate input
if (x1 in [0,1]) and (x2 in [0,1]):
    result = mcculloch_pitts_andnot(x1, x2)
    print("Output (x1 AND NOT x2):", result)
else:
    print("Invalid input! Please enter only 0 or 1.")
```
