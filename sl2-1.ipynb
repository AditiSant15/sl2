```python
import numpy as np
import matplotlib.pyplot as plt

# Input range
x = np.linspace(-10, 10, 200)

# 1. Sigmoid
def sigmoid(x):
    return 1 / (1 + np.exp(-x))

# 2. Tanh
def tanh(x):
    return np.tanh(x)

# 3. ReLU
def relu(x):
    return np.maximum(0, x)

# 4. Leaky ReLU
def leaky_relu(x):
    return np.where(x > 0, x, 0.1 * x)

# 5. ELU
def elu(x, alpha=1):
    return np.where(x > 0, x, alpha * (np.exp(x) - 1))

# 6. Step Function
def step(x):
    return np.where(x >= 0, 1, 0)

# 7. Bipolar Sigmoid
def bipolar_sigmoid(x):
    return (2 / (1 + np.exp(-x))) - 1

# 8. Softmax (for visualization only)
def softmax(x):
    exp_x = np.exp(x - np.max(x))   # stability trick
    return exp_x / np.sum(exp_x)

# Create plots
plt.figure(figsize=(18,16))

plt.subplot(3, 3, 1)
plt.plot(x, sigmoid(x))
plt.title("Sigmoid")
plt.grid()

plt.subplot(3, 3, 2)
plt.plot(x, tanh(x))
plt.title("Tanh")
plt.grid()

plt.subplot(3, 3, 3)
plt.plot(x, relu(x))
plt.title("ReLU")
plt.grid()

plt.subplot(3, 3, 4)
plt.plot(x, leaky_relu(x))
plt.title("Leaky ReLU")
plt.grid()

plt.subplot(3, 3, 5)
plt.plot(x, elu(x))
plt.title("ELU")
plt.grid()

plt.subplot(3, 3, 6)
plt.plot(x, step(x))
plt.title("Step Function")
plt.grid()

plt.subplot(3, 3, 7)
plt.plot(x, bipolar_sigmoid(x))
plt.title("Bipolar Sigmoid")
plt.grid()

plt.subplot(3, 3, 8)
plt.plot(x, softmax(x))
plt.title("Softmax")
plt.grid()

plt.tight_layout()
plt.show()
```
